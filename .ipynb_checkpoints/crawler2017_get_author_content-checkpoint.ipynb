```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://baike.baidu.com/item/%E5%AF%B9%E5%81%B6%E7%A9%BA%E9%97%B4'
```

```python
#loc = url
#seen = set()
#res = requests.get(url)
kv = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; WOW64; rv:44.0) Gecko/20100101 Firefox/44.0"}
res = requests.get(url, headers = kv, allow_redirects=False)
html = res.content
html_doc = str(html, 'utf-8')

#new_url = res.headers["Location"]
#html = res.content
print(html_doc)
```

```python
soup = BeautifulSoup(html_doc)
```

```python
para = soup.select('.para')
```

```python
print(para)
```

```python
for content in para:
    print(content.text)
```

```python
author = soup.select('.pro-content')
print(author)
```

```python
author = soup.select('.pro-content-title h4')[0].text
author_title = soup.select('.pro-content-title span')[0].text
institute = soup.select('.pro-content p')[0].text
print(author, author_title, institute)
```
